To facilitate and enhance better clarification and visualization plot, we will be regrouping the entries in the "Occupation" column  due to the large unique values (1100) in occupation column.

```python
# Let's regroup the entries in the Occupation column into a new classification

# Let's define a functiont to group the occupation value

def group_occupation (occupation):
    if  occupation == "Unemployed":
        return "Unemployed"
    elif occupation == "Child":
        return "Child"
    elif occupation =="Student":
        return "Student"
    elif occupation == "University Student":
        return "University Student"
    elif occupation  == "PhD Student":
        return "PhD Student"
    elif occupation.startswith("Retired"):
        return "Retired"
    else :
        return "Employed"
```

```python
# Let's create a new column for occupation classification

df["Occupation_Group"] = df["Occupation"].apply(group_occupation)
```

## Examination of Age Distribution (Age Pyramid) of the Population

```python
# Here, we are plotiting the Age Pyramid

# Calculate the min and max age values
min_age = int(df["Age"]. min())
max_age = int(df["Age"]. max() + 0.1)

# Let's set the bin size

bin_size = 5
bins = range(min_age, max_age + bin_size, bin_size)

# Let's create the age groups
df["Age_Groups"] = pd.cut(df["Age"], bins = bins, labels = [f'{i}-{i+bin_size-1}' for i in bins[: -1]])

# Let's group by Age group and Gender and then calculate the total count
Age_Groups = df.groupby(["Age_Groups", "Gender"]).size().unstack()

# Let's seperate female and male Dataframes
females = Age_Groups["Female"].sort_index(ascending = True)
males = Age_Groups["Male"].sort_index()

# plotting the Age Pyramid
fig, ax = plt.subplots(figsize =(8,6))   #(15, 12)

# Plotting Female and Male graph
ax.barh(males.index, males, height = 0.5, align = "center", color = "green", alpha = 0.7, left = 0, label = "Male")
ax.barh(females.index, -females, height = 0.5, align = "center", color = "red", alpha = 0.7, label = "Female")

# Setting the x-axis limits and ticks
max_count = max(abs(males).max(), abs(females).max())
ax.set_xlim([-max_count * 1.1, max_count * 1.1])  
ax.set_xticks(range(-int(max_count * 1.1), int(max_count * 1.1) + 1, int(max_count * 0.1)))

                
# Add vertical line at x = 0
ax.axvline(x=0, linestyle= '-', color = "black", alpha=0.5)
                
# Add Labels and  titles
                
ax.set_xlabel('Frequency')
ax.set_ylabel('Age Group')
ax.set_title(' Population Pyramid (Age Distribution)')
plt.xticks(ticks = [-500, -400, -300,-200, -100, 0, 100, 200, 300, 400, 500],labels = ["500", "400", "300", "200", "100", "0", "100", "200", "300", "400","500"])
    
# Add legend
ax.legend(loc = "upper right")

# Display plot                
plt.show()
                
```

The young people have a slightly lower number of population compared to middle aged people. The number start declining as the age group keeps increasing. The lowest age group for female is 100-104 why the lowest age group for male is 105-109.

```python
# Calculate the birth rate of the population in the census data
# Get counts of births for instance ages 0
# Get counts of all women between the ages 0f 25-29
# Calculate the birth rate

Number_of_Births = (df["Age"]== 0).sum()
Count_of_Women = ((df["Gender"]=="Female") & (df["Age"] >= 25) & (df["Age"] <= 29)).sum()
Birth_Rate = ((Number_of_Births/Count_of_Women)* 1000).round(2)
Birth_Rate
```

### Determination of the Commuters

The following are the commuters who are likely to commute to nearby cities, those with occupation, PhD Student, University Student, Science writer, Scientist,clinical(histocompatibilty and immunogenetics),Scientist audiological, Scientist, water quality, Scientist marine, Scientist research (life sciences), Scientist biomedical,Soil scientist, Scientist physiological, Scientist research(medical), Scientist, production/process developement, Librarian public, Librarian academic, Education officer community,Education administrator, Further education lecturer, Lecturer Further Education, Research Scientist (medical), Research officer Trade union, Research scientist (maths), Higher Education careers advisers, Journalist newspaper, Jounalist margazine, Newspaper journalist, Educational officer environmental, Educational psychologist, Clinical research associate, Teaching labouratory technical, also we have the Religion group who will likely commute i.e Methodist, Muslim, Christian, Sikh, Jewish, Agnostic and Bahai.

```python
# To get the Population of persons by their Occupation who are most likely to commute to nearby Cities 
# Let's define a function to group the Occupantion columninto commuters and non-commuters

def Commuters_Occupation(occupation):
    
    if occupation =="University Student":
        return "University Student"
    elif occupation == "PhD  Student":
        return "PhD Student"
    elif occupation.startswith("Scientist"):
        return "Scientist"
    elif occupation.startswith("Education"):
        return "Education"
    elif occupation.startswith("Lecturer"):
        return "Lecturer"
    elif occupation.startswith("Research"):
        return "Research"
    elif occupation.startswith("Librarian"):
        return "Librarian"
    elif occupation.startswith("Journalist"):
        return "Journalist"
    elif occupation.startswith(" Higher"):
        return "Higher"
    elif occupation.startswith("Educational"):
        return "Educational"
    elif occupation.startswith("Newspaper"):
        return "Newspaper"
    elif occupation.startswith("Further"):
        return "Further"
    elif occupation.startswith("Clinical"):
        return "Clinical"
    else:
        return "Non Commuter"
    
Commuters_Group = df["Occupation"].apply(Commuters_Occupation)
```

```python
Commuters_Group.unique()
```

```python
# Calculate the commuters entries in commute group, indivdiduals by occupation in the town who are likely to commute to nearby cities

Commuters_Occup = df.loc[Commuters_Group !="Non Commuter"].value_counts().sum()
Commuters_Occup
```

```python
# To get the Population of persons by their Religion affiliation  who are most likely to commute to nearby Cities.
# Let's define a function to group the religion feature into commuters and non-commuters

def Commuters_Religion(religion):
    
    if religion =="Methodist":
        return "Commuter"
    elif religion == "Muslim":
        return "Commuter"
    elif religion == "Christian":
        return "Commuter"
    elif religion == "Sikh":
        return "Commuter"
    elif religion == "Jewish":
        return "Commuter"
    elif religion == "Agnostic":
        return "Commuter"
    elif religion == "Bahai":
        return "Commuter"
    
    else:
        return "Non Commuter"
    
Commuters_Group_Religion = df["Religion"].apply(Commuters_Religion)
```

```python
# Display the unique values in Commuters_Group_Religion

Commuters_Group_Religion.unique()
```

```python
# Calculate the commuters entries in commute group, indivdiduals by Religion Affiliatin in the town who are likely to commute to nearby cities

Commuters_reli = df.loc[Commuters_Group_Religion != "Non Commuter"].value_counts().sum()
Commuters_reli
```

```python
# Calculate by Occupation and Religion the total number of persons who are most likely to commute to nearby cities

Total_number_of_commuters = Commuters_Occup + Commuters_reli
Total_number_of_commuters
```

From the above given information, there are 4048 total number of commuters in the population.

## Determination of Migration Rate

```python
# Calculate the population's immigration Levelin the Town
# The people that are most likely to immigrate are;
# Relationship to Head of House "None"
# Relationship to Head of House "Visitor"
# Relationship to Head of House "Lodger" 
# Relationship to Head of House "Head" and Occupation "University Student"
# Number of immigrants in the population

Number_of_immigrants = (df[df["Relationship_to_Head_of_House"] == "None"].value_counts().sum()) + (df[df["Relationship_to_Head_of_House"] == "Visitors"].value_counts().sum()) + (df[df["Relationship_to_Head_of_House"] == "Lodger"].value_counts().sum()) + (df[(df["Relationship_to_Head_of_House"] == "Head") & (df["Occupation"]== "University Student")].value_counts().sum()) 

Number_of_immigrants
```

```python
# Calculate the population's emmigration Levelin the Town

# The people that are most likely to emmigrate are;
# Relationship to Head of House "None" and Occupation "University Student"
# Relationship to Head of House "None" and Occupation "PhD Student"
# Relationship to Head of House "Head" and Occupation "University Student"
# Relationship to Head of House "Lodger" and Marital Status "Divorced"
# Relationship to Head of House "Lodger"  and  Occupation "Student"
# Relationship to Head of House "Lodger " and Occupation "University Student"
# Marital Status as "Divorced"

# Number of emmigrants in the population
Number_of_Emmigrants = (df[(df["Relationship_to_Head_of_House"] == "None") & (df["Occupation"] == "University Student")].value_counts().sum() + df[(df["Relationship_to_Head_of_House"] == "None") & (df["Occupation"] == "PhD Student")].value_counts().sum() +     df[(df["Relationship_to_Head_of_House"] == "Head") & (df["Occupation"] == "University Student")].value_counts().sum() +     df[(df["Relationship_to_Head_of_House"] == "Lodger") & (df["Marital_Status"] == "Divorced")].value_counts().sum() +     df[(df["Relationship_to_Head_of_House"] == "Lodger") & (df["Occupation"] == "Student")].value_counts().sum() +     df[(df["Relationship_to_Head_of_House"] == "Lodger") & (df["Occupation"] == "University Student")].value_counts().sum() +     df[df["Marital_Status"] == "Divorced"].value_counts().sum() )
Number_of_Emmigrants
```

```python
# Calculate the Migration Rate

Migration_Rate = (((Number_of_immigrants - Number_of_Emmigrants)/Population) *1000). round(2)
Migration_Rate
```
